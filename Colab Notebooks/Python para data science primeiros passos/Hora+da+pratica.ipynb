````python
n1 = int(input('Digite um número: '))
operador = input('Digite o operador: ')
n2 = int(input('Digite outro número: '))

resultado = 0

if operador == '+':
    resultado = n1 + n2
    print(f'A soma entre {n1} e {n2} é {resultado}')

elif  operador == '-':
    resultado = n1 - n2
    print(f'A subtração entre {n1} e {n2} é {resultado}')

elif operador == '*':
    resultado = n1 * n2
    print(f'A multiplicação entre {n1} e {n2} é {resultado}')

elif operador == '/':
    resultado = n1 / n2
    print(f'A divisão entre {n1} e {n2} é {resultado}')

if resultado % 2 == 0:
    print('O número é par')

else:
    print('O número é ímpar')

if resultado % 1 == 0:
    print('O número é inteiro')

else:
    print('O número é decimal')

if resultado > 0:
    print('O número é positivo')

else:
    print('O número é negativo')
````

Para o diesel, se a quantidade comprada for até 15 litros, o desconto será de 3% por litro. Caso contrário, será de 5% por litro. O preço do litro de diesel é R\$ 2,00 e o preço do litro de etanol é R\$ 1,70. Escreva um programa que leia a quantidade de litros vendidos e o tipo de combustível (E para etanol e D para diesel) e calcule o valor a ser pago pelo cliente. Tenha em mente algumas dicas:

- O do valor do desconto será a multiplicação entre preço do litro, quantidade de litros e o valor do desconto.
- O valor a ser pago por um cliente será o resultado da multiplicação do preço do litro pela quantidade de litros menos o valor de desconto resultante do cálculo.


````python
from re import I
combustivel = input('Digite o tipo de combustível E p/ Etanol ou D p/ Diesel: ')
litros = int(input(f'Digite a quantidade de litros de {combustivel}: '))
dieselPorLitro = 2
etanolPorLitro = 1.7
descontoEtanol = 0.02
descontoDiesel = 0.03
descontoEtanol2 = 0.04
descontoDiesel2 = 0.05

if combustivel == 'E':
    if litros <= 15:
        valor = (etanolPorLitro * litros) - (etanolPorLitro * litros * descontoEtanol)
        print(f'O valor a ser pago é R${valor: .2f}')

    else:
        valor = (etanolPorLitro * litros) - (etanolPorLitro * litros * descontoEtanol2)
        print(f'O valor a ser pago é R${valor: .2f}')

elif combustivel == 'D':
    if litros <= 15:
        valor = (dieselPorLitro * litros) - (dieselPorLitro * litros * descontoDiesel)
        print(f'O valor a ser pago é {valor: .2f}')

    else:
        valor = (dieselPorLitro * litros) - (dieselPorLitro * litros * descontoDiesel2)
        print(f'O valor a ser pago é {valor: .2f}')
````

- Para variação acima de 20%: bonificação para o time de vendas.

Dica: para calcular o percentual de crescimento usamos a seguinte equação: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
bacDia = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]
lista = []
for i in range(1, len(bacDia)):
    amostra_atual = bacDia[i]
    amostra_passada = bacDia[i - 1]
    percentual = 100 * (amostra_atual - amostra_passada) / amostra_passada
    lista.append(percentual)
    print(f'O percentual de crescimento do dia {i} é {percentual:.2f}%')
````
