#Operadores numericos

```python
q_seguranca = 5
s_seguranca = 3000

q_docente = 16
s_docente = 6000

q_diretoria = 1
s_diretoria = 12500
```

```python
q_funcionarios = q_seguranca + q_docente + q_diretoria
q_funcionarios
```

```python
s_diferenca = s_diretoria - s_seguranca
s_diferenca
```

```python
media = (q_seguranca * s_seguranca + q_docente * s_docente + q_diretoria * s_diretoria) / (q_seguranca + q_docente + q_diretoria)
media
```
